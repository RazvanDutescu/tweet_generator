```python
with open('twitter_credential.json') as cred:
    info = json.load(cred)
    customer_key = info['cust_key']
    customer_secret = info['cust_secret']
    access_key = info['access_key']
    access_secret = info['access_secret']
```

```python
def get_tweets(name):
    auth = tweepy.OAuthHandler(customer_key, customer_secret)
    auth.set_access_token(access_key, access_secret)
    api = tweepy.API(auth)
    
    all_tweets = []
    
    new_tweets = api.user_timeline(screen_name = name, count = 200)
    
    all_tweets.extend(new_tweets)
    
    old_tweets = all_tweets[-1].id -1
    
    while len(new_tweets)  != 0: #&amp;amp;amp;amp;gt; 0:
        new_tweets = api.user_timeline(screen_name = name, count = 200, max_id = old_tweets)
        all_tweets.extend(new_tweets)
        old_tweets = all_tweets[-1].id -1
        print('...%s tweets have been downloaded so far' % len(all_tweets))
        
        outtweets = [[ tweet.text] for tweet in all_tweets]
        
        with open(name + '_tweets.csv' , 'w', encoding = 'utf8') as f:
            writer = csv.writer(f)
            writer.writerow(['text'])
            writer.writerows(outtweets)
    return all_tweets
```

```python
def get_user_tweets( username, since_id=None):
        """
        Download all tweets for an user
        Max is around 3200 tweets
        """
        
        auth = tweepy.OAuthHandler(customer_key, customer_secret)
        auth.set_access_token(access_key, access_secret)
        api = tweepy.API(auth)
        tweets = []
        if since_id:
            cursor = tweepy.Cursor(api.user_timeline, screen_name=username, since_id=since_id)
        else:
            cursor = tweepy.Cursor(api.user_timeline, screen_name=username)

        for item in cursor.items():
            if not item.retweeted:
                tweets.append(item)

        return tweets 
```
